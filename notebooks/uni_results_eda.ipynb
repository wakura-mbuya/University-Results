# Exploratory Data Analysis and Visualization
<p>The Exploratory data analysis will help me get insights of the dataset. The data used in this project was collected from the Faculty of Engineering and Faculty of Educational Sciences students in 2019 from a university in Cyprus. The dataset can be found on Kaggle through <a href="https://www.kaggle.com/datasets/csafrit2/higher-education-students-performance-evaluation">this link</a></p>
<p>The dataset is made up of the following features:</p>
<ul>
    <li><b>STUDENTID</b></li>
    <li><b>AGE </b>- (1: 18-21, 2: 22-25, 3: 26 and above)</li>
    <li><b>GENDER</b> - (1: Female, 2: Male)</li>
    <li><b>HS_TYPE </b>- Graduated high school type (1: Private, 2: State, 3: Other)</li>
    <li><b>SCHOLARSHIP</b> - Scholarship type (1: None, 2: 25%, 3: 50%, 4: 75%, 5: Full)</li>
    <li><b>WORK</b> - Additional work (1: Yes, 2: No)</li>
    <li><b>ACTIVITY</b> - Regular artistic or sports activity (1: Yes, 2: No)</li>
    <li><b>PARTNER</b> - Does the student have a partner(1: Yes, 2: No)</li>
    <li><b>SALARY</b> - Total salary if available (1: 135 - 200 USD, 2: 201-270 USD, 3: 271-340 USD, 4: 341-410 USD, 5: 410 and above) </li>
    <li><b>TRANSPORT</b> - Transportation to the university (1: Bus, 2: Private car/taxi, 3: bicycle, 4: Other)</li>
    <li><b>LIVING</b> - Accomodation type in Cyprus (1: rental, 2: Dormitory, 3: With family, 4: Other)</li>
    <li><b>MOTHER_EDU</b> - Mother's Education (1: Primary school, 2: Secondary School, 3: high school, 4: University, 5: MSc, 6: Ph.D)</li>
    <li><b>FATHER_EDUC</b> - Father's education (1: Primary school, 2: Secondary School, 3: high school, 4: University, 5: MSc, 6: Ph.D)</li>
    <li><b>#_SIBLINGS</b> - Number of siblings</li>
    <li><b>KIDS</b> - Number of children the student has</li>
    <li><b>MOTHER_JOB</b> - Mother's Occupation </li>
    <li><b>FATHER_JOB</b> - Father's Occupation </li>
    <li><b>STUDY_HRS</b> - Weekly study hours (1: None, 2: &lt;5 hours, 3: 6-10 hours, 4: 11-20 hours, 5: more than 20 hours)</li>
    <li><b>READ_FREQ</b> - Reading frequency for non-scientific books/journals (1: None, 2: Sometimes, 3: Often)</li>
    <li><b>READ_FREQ_SCI</b> - Reading frequency of scientific books/journals (1: None, 2: Sometimes, 3: Often)</li>
    <li><b>ATTEND_DEPT</b> - Attendance to the seminars/conferences related to the department (1: yes, 2: No)</li>
    <li><b>IMPACT</b> - Impact of your projects/activities on your success (1: positive, 2: negative, 3: neutral)</li>
    <li><b>ATTEND</b> - Attendance to classes (1: always, 2: sometimes, 3: never)</li>
    <li><b>PREP_STUDY</b> - Preparation to midterm exams 1 (1: alone, 2: with friends, 3: not applicable)</li>
    <li><b>PREP_EXAM</b> - Preparation to midterm exams 2 (1: closest date to the exam, 2: regularly during the semester, 3: never)</li>
    <li><b>NOTES</b> - Taking notes in classes (1: never, 2: sometimes, 3: always)</li>
    <li><b>LISTENS</b> - Listening in classes (1: never, 2: sometimes, 3: always)</li>
    <li><b>LIKES_DISCUSS</b> - Discussion improves the student's interes and success in the course (1: never, 2: sometimes, 3: always)</li>
    <li><b>CLASSROOM</b> - Flip classroom (1: not useful, 2: useful, 3: not applicable)</li>
    <li><b>CUML_GPA</b>Cumulative grade point average in the last semester out of 4.00 (1: &lt2.00, 2: 2.00-2.49, 3: 2.50-2.99, 4: 3.00 - 3.49, 5: above 3.49) </li>
    <li><b>EXP_GPA</b>Expected cumulative grade point average in the graduation out of 4.00 (1: &lt2.00, 2: 2.00-2.49, 3: 2.50-2.99, 4: 3.00 - 3.49, 5: above 3.49) </li>
    <li><b>COURSE ID</b>
    <li><b>GRADE</b> - Output Grade (0: Fail, 1: DD, 2: DC, 3: CC, 4: CB, 5: BB, 6: BA, 7: AA)</li>
</ol>

```python
def employment_label(row):
    '''Decode the work column'''
    if row['WORK'] == 1:
        return 'Employed'
    else:
        return 'Unemployed'
    
def study_hrs_label(row):
    '''Decode study hours column'''
    if row['STUDY_HRS'] == 1:
        return 'None'
    elif row['STUDY_HRS'] == 2:
        return 'Less than 5 hours'
    elif row['STUDY_HRS'] == 3:
        return '6-10 Hours'
    elif row['STUDY_HRS'] == 4:
        return '11 - 20 Hours'
    else:
        return 'More than 20 Hours'
```

```python
work_df.columns
```

```python
work_df=results[['WORK', 'STUDY_HRS']].copy()
work_df =work_df.groupby(['WORK', 'STUDY_HRS'])['STUDY_HRS'].sum().to_frame(name='count').reset_index()
work_df['emp_status'] = work_df.apply (lambda row: employment_label(row), axis=1)
work_df['study_hrs'] = work_df.apply (lambda row: study_hrs_label(row), axis=1)
work_df.drop(['WORK', 'STUDY_HRS'], inplace=True, axis=1)
work_df
# for readability, change High school type values from the coded value to some meaningful string
# highSkul_df.at[0, 'HS_TYPE'] = 'Private'
# highSkul_df.at[1, 'HS_TYPE'] = 'Public'
# highSkul_df.at[2, 'HS_TYPE'] = 'Other'

# I need to caclulate the averate weekly hours spent on studying
```
